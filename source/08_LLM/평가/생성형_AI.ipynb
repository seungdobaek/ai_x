# 다음의 생성형 ai를 렝체인으로 구현하시오
- 원하는 모델을 이용하여 나라를 입력하면 해당 나라의 가장 유명한 음식을 추천하는 렝체인을 구현 (food_chain)
- 음식을 입력하면 레시피를 생성하는 렝체인을 구현 (recipe_chain)
- 위의 두 체인을 연결하여, 나라 이름을 받아, 해당 나라의 가장 유명한 음식의 레시피를 생성하는 렝체인 프로그램을 구현하시요

```python
# 사용할 패키지 로드
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# llm 모델 생성
llm = ChatOllama(model="llama3.2:1b")
```

## 원하는 모델을 이용하여 나라를 입력하면 해당 나라의 가장 유명한 음식을 추천

```python
# 메세지 생성
message = ChatPromptTemplate([
    ('system', 'You are an expert cook helping to guide recipe'),
    ('human', 'Please provide me one famous food in England? Give me the name of the food only'),
    ('ai', 'fish&chips'),
    ('human', 'Please provide me one famous food in Korea? Give me the name of the food only'),
    ('ai', 'kimchi'),
    ('human', 'Please provide me one famous food in Thailand? Give me the name of the food only'),
    ('ai', 'pho'),
    ('human', 'Please provide me one famous food in China? Give me the name of the food only'),
    ('ai', 'dim sum'),
    ('human', 'Please provide me one famous food in Italy? Give me the name of the food only'),
    ('ai', 'spaghetti'),
    ('human', 'Please provide me one famous food in United States? Give me the name of the food only'),
    ('ai', 'hambuerger'),
    ('human', 'Please provide me one famous food in Japan? Give me the name of the food only'),
    ('ai', 'sushi'),
    ('human', 'Please provide me one famous food in {country}? Give me the name of the food only')
])
```

```python
country = input("Please type the country : ")
output_parser = StrOutputParser()
food_chain =  message | llm | output_parser
food_chain.invoke('country')
```

## 음식을 입력하면 레시피를 생성하는 렝체인을 구현

```python
# 메세지 생성
message = ChatPromptTemplate([
    ('system', 'You are an expert cook helping to guide recipe'),
    ('human', 'Please provide me the receipe of {food}')
])

food = input("Please type food : ")
output_parser = StrOutputParser()
receipe_chain = message | llm | output_parser
receipe_chain.invoke({'food':food})
```

## 위의 두 렝체인 합치기

```python
# 방식1
country = input("Please type the country : ")
added_chain = food_chain | receipe_chain
added_chain.invoke({'country':country})
```

```python
# 방식2
from langchain_core.runnables import RunnablePassthrough
added_chain = {'country': RunnablePassthrough()} | {'food':food_chain} | receipe_chain
country = input("Please type the country : ")
added_chain.invoke({'country':country})
```
