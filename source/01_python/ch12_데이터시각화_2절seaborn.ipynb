```python
# https://stackoverflow.com/questions/30490740/move-legend-outside-figure-in-seaborn-tsplot
sns.set(style='whitegrid', 
        context='notebook',
        palette='Set1',
        rc={'figure.figsize':(10,3)})

# sns.set (미적속성)을 설정하면 기존의 설정들이 reset되기에 한글설정을 다시해야만 한다
plt.rc('font', family='Malgun Gothic') # 윈도우즈
#plt.rc('font', family='Apple Gothic') # 맥
plt.rc('axes', unicode_minus=False) # 축의 - 깨짐을 방지

ax = sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', palette='hsv')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set(title='상관계수가 높은 두 변수',
       ylabel='꽃잎너비')
ax.set_xlabel('꽃잎 길이', fontsize=20)
ax.set_xticks([1, 3, 5, 7])
plt.show()
```

## 2.3 seaborn 함수로 그래프 그리기
### 1) 관계형 그래프
- x축이 수치인 그래프
- scatterplot, lineplot, relplot(서브플롯제공) : x가 수치형

```python
# 종별 다른색상과 다른 스타일로 scatter
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species')
plt.show()
```

```python
# 종별 다른색상과 다른 스타일로 lineplot
sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species')
plt.show()
```

```python
# 종별 lineplot을 서브플롯으로
sns.relplot(data=iris, x='petal_length', y='petal_width', hue='species', 
            kind='line', # 서브플롯으로 그려질 그래프 종류 (기본값:scatter)
            col='species', # col 단위로 서브플롯 표현
            #col_wrap=2, # 한줄에 표현될 서브플롯 수
            #facet_kws={'sharey':False}
           )
plt.show()
```

```python
# 종별 scatter plot으로 서브플롯으로
g = sns.relplot(data=iris, x='petal_length', y='petal_width', hue='species', 
            kind='scatter', # 서브플롯으로 그려질 그래프 종류 (기본값:scatter)
            col='species', # col 단위로 서브플롯 표현
            #col_wrap=2, # 한줄에 표현될 서브플롯 수
            #facet_kws={'sharey':False}
           )
# 크기 조정
g.fig.set_size_inches(10,3)
plt.show()
```

```python
# ax[0]에는 scatter, ax[1]에는 line
fig, ax = plt.subplots(ncols=2)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=ax[0])
sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=ax[1])
plt.show()
```

### 2) 범주형 그래프
- x축이 범주형인 그래프
- stripplot(무작위로 흩뿌림), swarmplot(겹치지 않음) - 점도표
- barplot, pointplot(막대대신 marker가 있는 선) - 바 계열
- countplot(bar는 y의 평균값이 높이, countplot은 x의 빈도수 높이)
- boxplot, violinplot, boxenplot     => catplot(서브플롯)


```python
# x축이 범주형일때의 산점도
sns.stripplot(data=iris, x='species', y='petal_width', hue='species', palette='hsv')
plt.show()
```

```python
# x축이 범주형일때의 산점도 (중첩되지 않게)
sns.swarmplot(data=iris, x='species', y='petal_width', hue='species', palette='hsv')
plt.show()
```

```python
# 내림차순으로 정렬된 iris 범주형 산점도
sns.swarmplot(data=iris.sort_values('petal_width', ascending=False),
             x='species', y='petal_width', hue='species', palette='hsv')
plt.show()
```

```python
# x 축의 순서를 임의로 정할 수 있음
sns.swarmplot(data=iris,
             x='species', y='petal_width', hue='species', palette='hsv',
             order=['versicolor', 'setosa', 'virginica'])
plt.show()
```

```python
# barplot : 각 범주별 평균값과 그에 대한 신뢰구간(신뢰구간제외: errorbar=None)
sns.barplot(data=iris, x='species', y='sepal_width',
            errorbar=None)
sns.lineplot(data=iris, x='species', y='sepal_width', color='k')
plt.show()
```

```python
sns.pointplot(data=iris, x='species', y='sepal_width')
plt.show()
```

### 5) 회귀 그래프

```python
corr = iris.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr, vmin=-1, vmax=1, annot=True, fmt='.3f', cmap='Greens', cbar=False)
plt.show()
```

```python
np.ones_like(corr, dtype=bool)
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
mask
```

```python
sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, annot=True, fmt='.3f', cmap='Greens', cbar=False)
plt.show()
```

```python
# 회귀 그래프: regplot(hue사용불가), lmplot(hue 사용가능)
sns.regplot(data=iris, x='petal_length', y='petal_width')
plt.show()
```

```python
sns.lmplot(data=iris, x='petal_length', y='petal_width', hue='species', col='species')
plt.show()
```
