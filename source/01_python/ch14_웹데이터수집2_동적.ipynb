## 2.2 페이징 처리
- 다음 뉴스 페이징 처리 : 원하는 keyword를 원하는 페이지만큼 조회

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

driver = webdriver.Chrome()
driver.get('https://www.daum.net/')
#driver.implicitly_wait(0.5) # 최대 0.5초 동안 대기
time.sleep(0.5)
driver.find_element(By.CLASS_NAME, 'tf_keyword').click()
query = input('검색할 단어는?')
driver.find_element(By.CSS_SELECTOR, 'input[type="text"]').send_keys(query)
driver.find_element(By.CLASS_NAME, 'btn_ksearch').click()
time.sleep(2)

# 뉴스탭 이동
driver.find_element(By.LINK_TEXT, '뉴스').click()

#뉴스제목과 link를 list에 추가
news_list = []
pages = 3
for page in range(1, pages+1):
    strongs = driver.find_elements(By.CSS_SELECTOR, 'div.item-title > strong')
    # print(len(strongs))
    for strong in strongs:
        a = strong.find_element(By.TAG_NAME, 'a')
        title = a.text
        link = a.get_attribute('href')
        #print(title, link)
        news_list.append([title,link])
    page_nav = driver.find_element(By.CSS_SELECTOR, 'div.inner_paging')
    page_nav.find_element(By.LINK_TEXT, str(page+1)).click()
    time.sleep(2)
driver.close()

pd.DataFrame(news_list, columns=['title', 'link'])
```

- 설화.txt 읽어 300자 단위로 짤라 list에 넣어 동적웹크롤링 한후, 결과 출력 (설화_맞춤법후.txt)

```python
with open('data/ch14_설화.txt', 'r', encoding='utf-8') as f:
    temp_text = f.read()
ready_text = [] #300자씩 나눠진 text
text = temp_text
while(len(text) > 300):
    temp = text[:300]
    last_dot_index = temp.rfind('\n')
    ready_text.append(text[:last_dot_index])
    text = text[last_dot_index+1 : ]
ready_text.append(text)
# print(ready_text)
# print([len(ready) for ready in ready_text])

driver = webdriver.Chrome()
driver.get('https://www.naver.com')
time.sleep(1)
elem = driver.find_element(By.ID, 'query')
elem.send_keys('맞춤법 검사기')
elem.send_keys(Keys.RETURN)
time.sleep(1)

textarea = driver.find_element(By.CSS_SELECTOR, 'textarea.txt_gray')
btn = driver.find_element(By.CLASS_NAME, 'btn_check')
results = '' #맞춤법 검사 완료된 test
for i, ready in enumerate(ready_text):
    print(f'검사중...{i+1}/{len(ready_text)}')
    textarea.clear()
    #textarea.send_keys(Keys.CONTROL, 'a') #ctrl-a
    textarea.send_keys(ready)
    btn.click()
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    result = soup.select_one('p._result_text.stand_txt').text
    results += result + '\n'
    time.sleep(1)
driver.close()
```

```python
with open('data/설화_맞춤법후.txt', 'w', encoding='utf-8') as f:
    f.write(results)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()
driver.get('http://nivelarena.co.kr/bbs/board.php?bo_table=cardlists')
time.sleep(2)
driver.find_element(By.CSS_SELECTOR, 'div#ip_sch > select').click()
time.sleep(1)
driver.find_element(By.CSS_SELECTOR, 'div#ip_sch > select > option:nth-child(4)').click()
time.sleep(2)
for i in range(7):
    driver.find_element(By.CSS_SELECTOR, 'button#more_btn').click()
    time.sleep(2)
driver.find_element(By.CSS_SELECTOR, 'div.gall_img > a').click()
time.sleep(1)
card_list=[]
while(not isAlertPresent(driver)):
    name = driver.find_element(By.CSS_SELECTOR, 'h2#subject').text
    type_ = driver.find_element(By.CSS_SELECTOR, 'h2#type').text
    image =  driver.find_element(By.CSS_SELECTOR, 'div#img > img').get_attribute('src')
    cost = driver.find_element(By.CSS_SELECTOR, 'table#tbl1 tr:nth-child(3) td:nth-child(2)').text
    rarity = driver.find_element(By.CSS_SELECTOR, 'table#tbl1 tr:nth-child(3) td:nth-child(4)').text
    power = driver.find_element(By.CSS_SELECTOR, 'table#tbl1 tr:nth-child(4) td:nth-child(2)').text
    hit = driver.find_element(By.CSS_SELECTOR, 'table#tbl1 tr:nth-child(4) td:nth-child(4)').text
    aff = driver.find_element(By.CSS_SELECTOR, 'table#tbl1 tr:nth-child(5) td:nth-child(2)').text
    keyword = driver.find_element(By.CSS_SELECTOR, 'table#tbl1 tr:nth-child(6) td:nth-child(2)').text
    card = {'name':name, 'type':type_, 'image':image, 'cost':cost, 'rarity':rarity, 'power':power, 'hit':hit, 'aff':aff, 'keyword':keyword}
    card_list.append(card)
    driver.find_element(By.CSS_SELECTOR, 'i.fa-arrow-circle-o-right').click()
    time.sleep(1)
```

```python
def isAlertPresent(driver):
    try:
        driver.switch_to.alert
        return True
    except:
        return False
```

```python
import pandas as pd
pd.DataFrame(card_list).to_csv('image_card/card.csv', index=False)
```

```python
card_df = pd.read_csv('image_card/card.csv')
card_df.head()
```

```python
def divide_type(row):
    split = row['type'].split('/')
    row['id'] = split[0].strip()
    row['type'] = split[1].strip()
    row['attr'] = split[2].strip()
    return row
card_df = card_df.apply(divide_type, axis=1);
```

```python
card_df.to_csv('image_card/card.csv', index=False)
card_df
```

```python
import requests
def save_image(row):
    'link의 이미지를 image_card/id, 확장자로 local에 저장'
    link = row['image']
    id_ = row['id']
    file_extension = link.split('.')[-1] #확장자
    quote_index = file_extension.find('?')
    if quote_index != -1:
        file_extension = file_extension[:quote_index]
    img = requests.get(link).content #이미지이기에 text가 안됨
    with open(f'image_card/{id_}.{file_extension}','wb') as f:
        f.write(img)
# save_image(card_df.iloc[0])
```
