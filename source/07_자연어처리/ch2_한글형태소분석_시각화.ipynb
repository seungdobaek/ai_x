## 3.1 HanNanum

```python
from konlpy.tag import Hannanum
hannanum = Hannanum(jvmpath=None,
                   # max_heap_size=256 # 기본값:1024 (1024MB)
                   )
# 입력된 텍스트를 가장 세밀하게 분석해서 형태소, 품사, 원형등의 후보군을 모두 반환
hannanum.analyze(text)
```

```python
# 형태소 분석 : morphs
print(hannanum.morphs(text))
```

```python
# 명사만 추출 : nouns
print(hannanum.nouns(text))
```

```python
print(hannanum.pos(text)) # ntags(품사갯수) 기본값: 9
```

```python
# 품사태그 : pos
print(hannanum.pos(text, ntags=22)) 
```

```python
# 퀴즈1 pos tag 결과에서 명사(NC, NQ, NB, NN)만 추출하기
tagged_text = hannanum.pos(text, ntags=22)
[token for token, tag in tagged_text if tag in ['NC', 'NQ', 'NB', 'NN']]
```

```python
# 퀴즈2 text에서 보통명사(NC)만 추출하기
[token for token, tag in tagged_text if tag=='NC']
```

```python
# 퀴즈3 text에서 형용사(PA)만 추출하기
[token for token, tag in tagged_text if tag=='PA']
```

```python
nouns = ['word1', 'word2', 'word1', 'word2']
' '.join(nouns)
```

```python
# emma 말뭉치에서 단어(특수문자 제외) 추출 -> 명사(NN, NNS, NNP, NNPS) 추출 -> 연결 -> 워드클라우드
from nltk.tokenize import RegexpTokenizer
from nltk.tag import pos_tag
# 단어들 추출
ret = RegexpTokenizer('[0-9a-zA-Z]{2,}')
emma_words = ret.tokenize(emma)
# pos tagging -> 명사 추출
emma_tags = pos_tag(emma_words)
noun_lists = [token for token, tag in emma_tags if tag in ['NN', 'NNS', 'NNP', 'NNPS']]
# 명사 연결
emma_noun = ' '.join(noun_lists)
print('워드 클라우드를 그릴 text 내용 :', emma_noun[:20])
```

```python
# 워드 클라우드
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wordc = WordCloud()
wordc.generate(emma_noun)
plt.imshow(wordc)
plt.show()
```

# 6. 워드 임베딩
- 단어간 벡터 계산 (밀집벡터의 형태)
- 단어간 거리 계산

```python
import requests
from bs4 import BeautifulSoup
from mecab import MeCab
rss_url = 'https://fs.jtbc.co.kr/RSS/economy.xml'
jtbc_response = requests.get(rss_url)
soup = BeautifulSoup(jtbc_response.content, 'xml')
item_elems = soup.select('item')
news = []
mecab = MeCab()
for item_elems in item_elems:
    title = item_elems.find('title').text
    description = item_elems.find('description').text
    total_text = title + ' ' + description
    noun_list = mecab.nouns(total_text)
    # Mecab의 경우: NNG(보통명사), NNP(고유명사)
    noun_list = [ token for token, tag in  mecab.pos(total_text) if tag in ['NNG', 'NNP']]
    news.append(noun_list)
print(news[:3])
```

```python
# 워드임베딩, Word2Vec(단어간 거리 계산)
# pip install gensim
from gensim.models import Word2Vec
model = Word2Vec(news, # 학습데이터 2차원
                window=10, #각 단어의 좌우 10개 단어만 학습 컨텐츠로 사용
                min_count=2, # 학습데이터에서 2회 이상 등장한 단어만 학습
                workers=-1, # 병렬처리할 core의 수 (-1:모든 코어를 사용)
                sg=1, # 소량의 데이터일 경우 중심단어로 주변단어를 예측하도록 학습
                )
```

```python
model.wv.most_similar('편의점', topn=5) # topn=10이 기본값
```

```python
# Word2Vec이 학습한 전체단어 어휘집
print(model.wv.index_to_key)
```

```python
# 유사한 단어를 출력
print('입력한 단어와 유사한 단어 10개를 출력합니다')
key = input()
if key in model.wv:
    print("{}와 유사한 단어들".format(key))
    print(model.wv.most_similar(key))
else:
    print("{} 단어가 훈련 데이터에 없음".format(key))
```
